## Step 2: Data Cleaning

```python
# Split the CarName column to extract the name of the car maker and label it as 'car_company'

CarName=pd.Series(car_prices['CarName'].str.split(" "))
car_prices['car_company'] = pd.DataFrame(CarName.str[0])

car_prices.head()
```

```python
#Check for names and spelling accuracy in 'car_company' 

car_prices['car_company'] = car_prices['car_company'].str.replace('-', ' ')
car_prices.car_company.unique()
```

### It is visible from the results above that there are some discrepancies in the names



```python
# Check the quality issues in the column car_company and replace with correct names

car_prices['car_company'] = car_prices['car_company'].str.replace("vw", "volkswagen")
car_prices['car_company'] = car_prices['car_company'].str.replace("vokswagen", "volkswagen")
car_prices['car_company'] = car_prices['car_company'].str.replace("toyouta", "toyota")
car_prices['car_company'] = car_prices['car_company'].str.replace("maxda", "mazda")
car_prices['car_company'] = car_prices['car_company'].str.replace("porcshce", "porsche")
car_prices['car_company'] = car_prices['car_company'].str.replace("Nissan", "nissan")
car_prices['car_company']
```

```python
# check and confirm for any further issues

car_prices['car_company'].value_counts()
```

```python
#  We can now drop the column 'CarName'

car_prices.drop(['CarName'], axis = 1, inplace = True)
car_prices.head()
```

## Step 4: Data Preparation for Analysis

#### Drop fields that are not required in predictive analysis

```python
# Drop car_ID since it is a serial no.

car_prices.drop(['car_ID'], axis = 1, inplace = True)
car_prices.head()
```

```python
# Drop `carwidth`,`curbweight`,`wheelbase`,`highwaympg` since they are highly correlated 

car_prices.drop(['carwidth','curbweight','wheelbase','highwaympg'], axis = 1, inplace = True)
car_prices.head()
```

#### Convert the Categorical data into numerical data

```python
car_prices['fueltype'] = car_prices.fueltype.map(dict(gas=1, diesel = 0))
car_prices['fueltype'].value_counts()
```

```python
car_prices['aspiration'] = car_prices.aspiration.map(dict(std=1, turbo = 0))
car_prices['aspiration'].value_counts()
```

```python
car_prices['doornumber'] = car_prices.doornumber.map(dict(two=1, four = 0))
car_prices['doornumber'].value_counts()
```

```python
car_prices['enginelocation'] = car_prices.enginelocation.map(dict(front=1, rear = 0))
car_prices['enginelocation'].value_counts()
```

### Dummy Variables

The variables having more than three levels can be converted to numerical variables by using something called `dummy variables`

```python
car_prices = pd.get_dummies(car_prices)
car_prices.head()
```

```python
# Check the columns for numerical presence

car_prices.info()
```

## Step 5:  Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(car_prices, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescaling the Features 

We will use MinMax scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['symboling','carlength', 'carheight','enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

#### Now the data is in the correct format and scale to proceed with further analysis

### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('price')
X_train = df_train
```

## Step 6: Building our model

We will be using the **LinearRegression function from SciKit Learn** for its compatibility with RFE (which is a utility from sklearn)

### RFE
Recursive feature elimination

```python
# Importing RFE and LinearRegression
# Since there are more than 60 variables to be analysed, we will use the automated feature elimination process (RFE)

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE 
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE for top 15 variables
rfe = rfe.fit(X_train, y_train)
```

```python
# Check the ranks
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Select the top 15 variables

col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# we can also check for multicollinearity at this stage to remove highly correlated variables

plt.figure(figsize = (20,10))  
sns.heatmap(X_train_rfe.corr(),annot = True)
```

#### Since it is visible from the heat map that `cylindernumber_two` and `enginetype_rotor` are highly corelated, we can drop `enginetype_rotor`.

```python
X_train_rfe = X_train_rfe.drop(['enginetype_rotor'], axis = 1)
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
#Check the summary of our linear model
print(lm.summary())
```

```python
X_train_rfe.columns
```

```python
# Check for the VIF values of the feature variables. 

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Step 7: Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_price = lm.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```
